```python
import numpy as a
n=a.arange(0.,5.,0.2)

plt.plot(n,n**2,'b--',label='^2')
plt.plot(n,n**2.2,'rs',label='^2.2')
plt.plot(n,n**2.5,'g^',label='^2.5')
plt.grid()
plt.show()
```
